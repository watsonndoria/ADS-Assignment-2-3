```python
# Import Library
import pandas as pd
```

```python
#Import csv
hr=pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

```python
#View Data
hr.head()
```

```python
#Show a breakdown of distance from home by job role and attrition.
(hr.groupby(["DistanceFromHome", "Department","Attrition"]).size()
 .reset_index(name='count').sort_values(by='Department', ascending=True))
```

```python
# avg monthly income
hr.MonthlyIncome=(hr.DailyRate*hr.StandardHours).mean()
```

```python
#Compare average monthly income by education and attrition.
hr.groupby(["MonthlyIncome", "Education","Attrition"]).size().reset_index(name='count')
```
